## Prepare SLM CA MCQ dataset in a given format to be used for training

```python
def read_in_chunks(file_path, chunk_size=10):
    with open(file_path, "r", encoding="utf-8") as f:
        while True:
            lines = []
            for _ in range(chunk_size):
                line = f.readline()
                if not line:
                    break
                lines.append(line.rstrip())

            if not lines:
                break  # no more data
            yield lines
```

```python
import os
import re
import json

instructions =  [] 
reports  = []
def read_text_files(root_folder):
    for dirpath, dirnames, filenames in os.walk(root_folder):
        mcq_counts = 0
        parent_folder = os.path.basename(dirpath)
        for filename in filenames:
            if filename.lower().endswith(".txt"):
                full_path = os.path.join(dirpath, filename)
                
                print(f"=== {full_path} ===")
                chunks = read_in_chunks(full_path)
                
                # Example usage
                for chunk in chunks:
                    #print("=== New Chunk ===")
                    instruction = {}
                    instruction["instruction"]= "Generate a Multiple Choice Question on the input context"
                    instruction["input"]=chunk[9]
                    instruction["output"]= f"Question:{re.sub(r'^\d+\.\s*', '', chunk[0])} \nOptions: \n{chunk[1]}  \n{chunk[2]}  \n{chunk[3]}  \n{chunk[4]} \n{chunk[7]}"
                    instructions.append(instruction)
                    print(instruction)
                    mcq_counts+=1
        if parent_folder=="training_ca_mcqs":
            continue; 
        report = {}
        report["month"] = f"{parent_folder}"
        report["mcq_count"] = mcq_counts
        reports.append(report)
        

# Example usage
read_text_files("../data/training_ca_mcqs")
print(f"{len(instructions)} Instructions generated")
print("Reports:",reports)
```

```python
from datetime import date, datetime

base_path = "../data/training_ca_mcqs"
today_date = date.today()
today_date_string = today_date.strftime("%Y%m%d")


with open(f"{base_path}/instruction-data-{today_date}.json", "w", encoding="utf-8") as f:
    json.dump(instructions, f, indent=4, ensure_ascii=False)


with open(f"{base_path}/instruction-data-{today_date}-report.json", "w", encoding="utf-8") as f:
    json.dump(reports, f, indent=4, ensure_ascii=False)
```

```python
from datetime import date
import json
import time

base_folder_path = "../data"
today_date = date.today()
today_date_string = today_date.strftime("%Y%m%d")
todays_news_folder =  f"{base_folder_path}/{today_date_string}"
print(todays_news_folder)


mcq_generation_instructions_file = f"{base_folder_path}/news_content/{today_date_string}/__{today_date_string}_ca_mcqs_generation_instructions.json"

with open(mcq_generation_instructions_file, "r", encoding="utf-8") as f:
    mcq_generation_instructions= json.load(f)  # indent=4 makes it pretty

print(f"mcq_generation_instructions: len={len(mcq_generation_instructions)}")
```

```python
def format_input(entry):
    instruction_text = (
        f"Below is an instruction that describes a task. "
        f"Write a response that appropriately completes the request."
        f"\n\n### Instruction:\n{entry['instruction']}"
    )

    input_text = f"\n\n### Input:\n{entry['input']}" if entry["input"] else ""

    return instruction_text + input_text
```

```python
from google import genai
# The client will read your API key from the environment variable

api_key="AIzaSyBtVF9I2ghTqIi7o5TZ29lP86DnlD-rHjk"
model="gemini-2.5-flash"
client = genai.Client(api_key=api_key)
```

```python
for index, real_nes_content in enumerate(mcq_generation_instructions):
    input_text = format_input(real_nes_content)

    response = client.models.generate_content(
        model=model,              # model of your choice
        contents=input_text
    )
    real_nes_content["output"] = response.text

    print(index,"=>",input_text,"\n")
    print("### Model Response\n",response.text)
    
    time.sleep(5)   # wait for 5 seconds
```

```python
mcq_generation_file = f"{base_folder_path}/news_content/{today_date_string}/__{today_date_string}_ca_mcqs_by_{model}.json"


with open(mcq_generation_file, "a", encoding="utf-8") as f:
    json.dump(mcq_generation_instructions, f, indent=4)  # indent=4 makes it pretty


print(f"{model} gemnerated MCQs and saved in {mcq_generation_file}")
```
